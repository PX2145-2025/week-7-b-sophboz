```python
import numpy as np
import matplotlib.pyplot as plt

# Number of steps
N = 1000

# Generate random steps
steps = np.random.choice([-1, 1], size=N)

# Compute position after each step
position = np.cumsum(steps)

# Plot the walk
plt.figure(figsize=(10, 5))
plt.plot(position)
plt.title("1D Random Walk (1000 steps)")
plt.xlabel("Step number (i)")
plt.ylabel("Position (x)")
plt.grid(True)
plt.show()
```

```python


# Parameters
N = 1000   # number of steps
M = 100    # number of walks

plt.figure(figsize=(10, 5))

# Generate and plot 100 walks
for _ in range(M):
    steps = np.random.choice([-1, 1], size=N)
    position = np.cumsum(steps)
    plt.plot(position, alpha=0.4)

plt.title("100 Random Walks (Each 1000 steps)")
plt.xlabel("Step number (i)")
plt.ylabel("Position (x)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Parameters
N = 1000   # number of steps
M1 = 10    # number of walkers for first part
M2 = 500   # number of walkers for second part


#part 1: 10 walkers, all on the same plot
plt.figure(figsize=(10, 5))

for _ in range(M1):
    steps = np.random.choice([-1, 1], size=N)
    position = np.cumsum(steps)
    plt.plot(position, alpha=0.7)

plt.title("10 Random Walkers (1000 steps each)")
plt.xlabel("Step number (i)")
plt.ylabel("Position (x)")
plt.grid(True)
plt.show()


#part 2: 500 walkers
# Create an array of shape (M2, N)
steps_all = np.random.choice([-1, 1], size=(M2, N))
positions_all = np.cumsum(steps_all, axis=1)  # cumulative sum along steps

# Compute mean position and mean squared position at each step
mean_position = np.mean(positions_all, axis=0)
mean_squared_position = np.mean(positions_all**2, axis=0)

# Print the values at the last step
print(f"Mean position at final step: {mean_position[-1]:.3f}")
print(f"Mean squared position at final step: {mean_squared_position[-1]:.3f}")

#part 3: Plot the mean position over time
plt.figure(figsize=(10, 5))
plt.plot(mean_position)
plt.title("Mean Position vs Step Number (500 walkers)")
plt.xlabel("Step number (i)")
plt.ylabel("Mean position x")
plt.grid(True)
plt.show()

#part 4: Plot the mean squared position over time
plt.figure(figsize=(10, 5))
plt.plot(mean_squared_position)
plt.title("Mean Squared Position vs Step Number (500 walkers)")
plt.xlabel("Step number (i)")
plt.ylabel("Mean squared position x")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
N = 1000   # number of steps
M = 500    # number of walkers
intermediate_steps = [10, 100, 1000]  # steps at which to plot histograms

# --- Simulate all walks ---
steps_all = np.random.choice([-1, 1], size=(M, N))
positions_all = np.cumsum(steps_all, axis=1)  # shape (M, N)

# --- Plot and save histograms ---
for step in intermediate_steps:
    plt.figure(figsize=(8, 5))

    # positions of all walkers at the chosen step
    positions_at_step = positions_all[:, step - 1]  # step index offset
    plt.hist(positions_at_step, bins=30, color='steelblue', edgecolor='black', alpha=0.7)

    plt.title(f"Histogram of Walker Positions at Step {step}")
    plt.xlabel("Position (x)")
    plt.ylabel("Number of Walkers")
    plt.grid(True)

    # Save to file
    filename = f"hist_step_{step}.png"
    plt.savefig(filename)
    plt.show()
    print(f"Saved histogram: {filename}")
```
